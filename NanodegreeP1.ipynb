```python
import numpy as np
import scipy
import scipy.stats
import pandas
import statsmodels.api as sm
import ggplot

from ggplot import *
```

```python
turnstile_weather = pandas.read_csv('D:/Documents/Udacity/IntroDataScience/turnstile_data_master_with_weather.csv')
```

```python
#print turnstile_weather
```

```python
def normalise(data):
    mean = data.mean()
    stdev = data.std()
    return (data - mean)/stdev
```

```python
features = turnstile_weather[['rain', 'meanwindspdi', 'mintempi', 'Hour']]
features = normalise(features)
```

```python
dummy_units = pandas.get_dummies(turnstile_weather['UNIT'], prefix='unit')
features = features.join(dummy_units)
```

```python
values = turnstile_weather['ENTRIESn_hourly']
```

```python
features.corr()
```

```python
res = sm.OLS(values,features).fit()
print res.summary()
```

```python
predictions = res.predict()
print predictions
```

```python
residuals = turnstile_weather['ENTRIESn_hourly'] - predictions
```

```python
def residual_plot(data):
    resiplot = ggplot(residuals, aes(residuals)) + geom_histogram(fill = 'blue', binwidth = 200) \
        + xlim(-7500, 7500) + xlab('Residuals') + ylab('Frequency') + ggtitle('Histogram of Residuals from The Linear Regression Model')
    return resiplot
```

```python
residual_plot(residuals)
```
